# Using the `WebLLM Chat` kernel to run built-in-ai LLMs as JupyterLite kernels

*If you are in the wiki view (static HTML rendering), click "edit" in the top right to open this [JupyterLite](https://jupyterlite.readthedocs.io/) notebook in your browser.*

The `WebLLM Chat` kernel (be sure it is selected in the kernel picker above) is derived from a JupyterLite kernel implemented by [Aleksandar Sekulovski](https://www.linkedin.com/in/aleksandar-s-486829128/) at https://github.com/aleks-s10/agent-client-kernel/tree/main/lite-kernel.  This version currently lives at https://github.com/wiki3-ai/built-in-ai-webllm-kernel with Python installer by [Jim White](https://www.linkedin.com/in/jamespaulwhite/) and Copilot.

It is a TS/JS kernel that wraps [`@built-in-ai/web-llm`](https://www.npmjs.com/package/@built-in-ai/web-llm) from https://github.com/jakobhoeg/built-in-ai which is an [AI SDK model provider](https://ai-sdk.dev/) for open models (using [WebLLM](https://webllm.mlc.ai/) which uses [WebGPU](https://www.w3.org/TR/webgpu/) to run [HuggingFace MLC-LLM models](https://huggingface.co/models?library=mlc-llm)) running directly in the browser by [Jakob Hoeg Mørk](https://www.linkedin.com/in/jakob-hoeg-moerk/).
