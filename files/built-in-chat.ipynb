# Chrome AI Built-in Prompt API Demo

*If you are in the wiki view (static HTML rendering), click "edit" in the top right to open this [JupyterLite](https://jupyterlite.readthedocs.io/) notebook in your browser.*

This notebook demonstrates using Chrome/Edge Built-In AI with the Prompt API (Gemini Nano) [https://developer.chrome.com/docs/ai/prompt-api](https://developer.chrome.com/docs/ai/prompt-api).

Source for this JupyterLite kernel, "Built-in AI Chat" (check that it is selected in the kernel picker above), is located at https://github.com/wiki3-ai/built-in-ai-chat-kernel.

**Requirements:**
- Chrome browser version 138+ 
- Enable experimental flags:
  - [chrome://flags/#prompt-api-for-gemini-nano](chrome://flags/#prompt-api-for-gemini-nano)
  - [chrome://flags/#optimization-guide-on-device-model](chrome://flags/#optimization-guide-on-device-model)
- Restart browser
  - [chrome://restart](chrome://restart)
- First use requires downloading the Gemini Nano model which may take a while because the model is around 2GB.

- Open JupyterLite (`edit` button in top right when in wiki view) and create a notebook selecting the "Built-in AI Chat" kernel.

This is a sample chat using the built-in-chat JupyterLite kernel for [`@built-in-ai/web-core`](https://www.npmjs.com/package/@built-in-ai/core) from https://github.com/jakobhoeg/built-in-ai which is an [AI SDK model provider](https://ai-sdk.dev/) which provides the Built In AI Prompt API .
